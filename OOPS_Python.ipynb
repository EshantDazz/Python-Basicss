# <div class='alert alert-success'> OOPS </div>

**POP** - We break the code into a set of functions and these functions call each other<br>
**OOP** - We break the code into a set of entities and these entities talk to each other. These entities have data and methods.

```python
class Complex:
    def __init__(self,a,b):
        self.real=a
        self.imag=b
    def printt(self):
        print(str(self.real)+"+"+str(self.imag)+"i")
    def add(self,c):
        self.real+=c.real
        self.imag+=c.imag
```

```python
a=Complex(10,20)
```

```python
a.printt()
```

```python
b=Complex(30,50)
```

```python
a.add(b)
```

```python
a.printt()
```

## <font color='green'> Encapsulation </font>

It refers to data hiding. Wrapping data members and methods into a single unit.<br>
Example when we change something inside the method then the user will directly call the method without knowing what happened inside and we can change the features of the method without hampering the user's desired output.

#### Example 1:

```python
class Eshant:
    def __init__(self):
        self.DS='Python'
        self.__DSA='JAVA'
    def merge(self):
        return self.DS+self.__DSA
```

```python
ob=Eshant()
print(ob.merge())
print(ob.DS)
```

```python
#ob.DSA
#ob.__DSA
```

The above mentioned codes will give `error` bcoz that is not the way to access `private` data members

```python
print(ob._Eshant__DSA ) #This is a way called name mangling to access the pricate members
```

```python
class Jangkilt:
    def __init__(c):
        c.course='ML'
        c.__code='Python'
    def display(c):
        return c.course+" hehe "+c.__code
    def set_code(c,x):
        c.__code=x
    def get_code(c):
        return c.__code
```

## <font color='green'> Class Methods and Static Members </font>

**Class Methods** - Used for modifying the class attributes and use class attributes<br>

```python
class school:
    name='Edmunds'
    def __init__(self,name,age):
        self.name=name
        self.age=age
    @classmethod
    def setname(cls,sname):
        cls.name=sname
```

```python
school.setname('KV')
print(school.name)

s=school('Eshant',15)
print(s.name)
```

**Static method** - These are for utility method. Methods which are not class specific and not object specific

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    @staticmethod
    def isAdult(age):
        return age>18

    def printDetails(self):
        print(self.name)
        print(self.age)
        print(self.isAdult(self.age))
```

```python
p=Person("Eshant",21)
```

```python
p.printDetails()
```

```python
p.isAdult(11)
```

## <font color='green'> Inheritence </font>

```python
class Person:
    def __init__(c,idd,name):
        c.idd=idd
        c.name=name

class Student(Person):
    def __init__(c,idd,name,score):
        super().__init__(idd,name)
        c.score=score
    def printDetails(c):
        print(c.idd)
        print(c.name)
        print(c.score)
```

```python
s=Student(100,'Eshant',9)
s.printDetails()
```

### Multilevel Inheritence

```python
class Person:
    def __init__(self,idd,name):
        self.idd=idd
        self.name=name
    
    def printDetails(self):
        print(self.idd)
        print(self.name)
        
class Employee(Person):
    def __init__(self,idd,name,sal):
        super().__init__(idd,name)
        self.sal=sal
    def printDetails(self):
        super().printDetails()
        print(self.sal)
        
class SalesEmployee(Employee):
    def __init__(self,idd,name,sal,si):
        super().__init__(idd,name,sal)
        self.salesIncome=si
    def printDetails(self):
        super().printDetails()
        print(self.salesIncome)
```

```python
ob=SalesEmployee(101,"Eshant",20000,30000)
```

```python
ob.printDetails()
```

```python
e=Employee(102,'Jayesh',10000)
```

```python
e.printDetails()
```

### Multiple Inheritence

```python
class Student:
    def __init__(self,sid,dept):
        self.sid=sid
        self.dept=dept

class Faculty:
    def __init__(self,eid,dept):
        self.eid=eid
        self.dept=dept
    def fun(self):
        print("Hello")

class PHDStudent(Student,Faculty):
    def __init__(self,idd,dept):
        super().__init__(idd,dept)  #This will take Student by default        
```

```python
ps=PHDStudent(101,'CSE')
```

```python
print(ps.sid)
print(ps.dept)

##      print(ps.eid)   --->this will give error

ps.fun()  #this will work bcoz only the constructor has gone for the first class written inside parenthesis
```

## <font color='green'> Polymorphism </font>

```python
def fun(a,b):
    print(a)
    print(b)
def fun(a,b,c):
    print(a)
    print(b)
    print(c)
```

```python
#fun(10,20)   ->This gives error
fun(20,40,50)
```

**Python has its own way of supporting static polymorphism** .**`In terms of number of parameters it doesn't support`**

```python
def fun(a,b):
    print(a)
    print(b)
```

```python
fun(20,30)
fun('Eshant','Das')
```

**There is a way to do polymorphism with respect to number of parameters**

```python
def fun(l):
    for x in l:
        print(x)
```

```python
fun([10,20,30])
print()
fun([19,20])
print()
fun(["eshant","das",101])
```

## <font color='green'> Method Overriding and Polymorphism </font>

Derived class has same name and parameter method as base class

```python
class Employee:
    def __init__(self,idd,name):
        self.idd=idd
        self.name=name
    def printDetails(self):
        print(self.idd)
        print(self.name)

class SalesEmployee(Employee):
    def __init__(self,idd,name,si):
        super().__init__(idd,name)
        self.salesIncome=si
        
    def printDetails(self):
        super().printDetails()
        print(self.salesIncome)
```

```python
e1=[Employee(101,'Eshant'),SalesEmployee(102,'Jayesh',20000)]
```

```python
for x in e1:
    print(x)
```

```python
for x in e1:
    x.printDetails()
    print()
```

## <font color='green'> Operator Overloading </font>

This is present in C++ and Python and not in Java

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def __add__(self,other):
        return (self.price+other.price, self.name+" "+other.name)
```

```python
p1=Product('bat',500)
p2=Product('ball',50)
print(p1+p2)
```

## <font color='green'> Python Abstract Class </font>

```python
from abc import ABC,abstractmethod
```

We cannot object of abstract classes, thats the rule every language follows

An abstract class can have abstract method and concrete methods as well

```python
class Polygon(ABC):
    def __init__(c,color):
        c.color=color
        
    @abstractmethod
    def printSides(c):
        pass
    
    def printColor(c):
        print(c.color)

class Triangle(Polygon):
    def __init__(c,color):
        super().__init__(color)
    def printSides(c):
        print("There are three sides")
```

```python
p=Triangle('blue')
```

```python
p.printColor()
```

```python
p.printSides()
```
