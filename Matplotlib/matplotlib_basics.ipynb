```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[5,7,2,6]
y=[2,7,1,9]
```

```python
fig,ax=plt.subplots(figsize=(5,8))

ax.plot(x,y)

ax.set(title="Simple Plot",
       xlabel='x-axis',
        ylabel='y-axis')

plt.show()
```

**Line Plot**

```python
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

**Scatter Plot**

```python
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
plt.show()
```

**Bar Plot**

```python
b={'Butter':20,'Jam':30,"Milk":50}
```

```python
fig,ax=plt.subplots()
ax.bar(b.keys(),b.values())
plt.show()
```

```python
fig,ax=plt.subplots()
ax.barh(list(b.keys()),list(b.values()))
plt.show()
```

**Histogram**

```python
x=np.random.randn(500)
```

```python
fig,ax=plt.subplots()
ax.hist(x);
```

**Subplots**

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0][0].plot(x)
ax[0][1].scatter(x,x**2)
ax[1][0].hist(x)
ax[1][1].bar(b.keys(),b.values());
```
