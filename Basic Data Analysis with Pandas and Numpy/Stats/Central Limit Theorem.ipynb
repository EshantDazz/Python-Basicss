```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(50)
x=np.random.randint(1,8,100)
```

```python
plt.hist(x);
```

### <font color='green'>**Before Applying Central Limit Theorem**</font>

```python
sns.displot(x,kind='kde');
```

### Taking 100 samples

```python
np.random.seed(50)
x=[np.mean(np.random.randint(1,8,100)) for i in range(100)]
plt.hist(x);
```

### Taking 1000 samples

```python
np.random.seed(50)
x=[np.mean(np.random.randint(1,8,100)) for i in range(1000)] #this time taking 1000 samples
plt.hist(x);
```

### <font color='green'>**After Applying Central Limit Theorem**</font>

```python
sns.displot(x,kind='kde');
```
