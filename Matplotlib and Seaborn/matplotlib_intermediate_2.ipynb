```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
```

```python
df.head()
```

```python
df['age'].plot.hist(bins=30);
```

```python
df['age'].plot.hist();
```

```python
over50=df[df['age']>50]
```

**pyplot method**

```python
over50.plot(kind='scatter',x='age',y='chol',c='target');  #seaborn would be a better alternative option
```

**oo method**

```python
fig,ax=plt.subplots(figsize=(15,8))
over50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax); 
```

**Full OO Method**

```python
fig,ax=plt.subplots(figsize=(15,7))
s=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'])
ax.set(title='Heart Disease and Cholesterol',xlabel='Age',ylabel='Cholesterol')
ax.legend(*s.legend_elements(),title='target');
```

```python
fig,ax=plt.subplots(figsize=(15,7))
s=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'])
ax.set(title='Heart Disease and Cholesterol',xlabel='Age',ylabel='Cholesterol')
ax.legend().set_visible(True)
```

**Alternative seaborn**

```python
import seaborn as sns
```

```python

sns.relplot(data=over50,x='age',y='chol',hue='target',height=8,aspect=1);
```
