# Bivariate analysis

# Relplot

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
plt.show()
```

1. Positive corelation , more total_bill more the tip
2. Density is more between 10 and 20
3. Data points are sparse beyond 25

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind='line')
plt.show()
```

```python
#Adding HUE
sns.relplot(data=tips,x="total_bill",y="tip",hue="day")
plt.show()
```

```python
#Adding col
sns.relplot(data=tips,x="total_bill",y="tip",col='day',col_wrap=2)
plt.show()
```

```python
#Adding col and hue together
sns.relplot(data=tips,x="total_bill",y="tip",hue='sex',col='day',col_wrap=2)
plt.show()
```

Inference
1. Males give more tips at night to impress girls

```python
sns.boxplot(data=tips,x='sex',y='tip',hue='day')
```

```python
sns.relplot(data=tips,x='size',y='tip',kind='line')
plt.show()
```

# Boxplot

Cateforical to numerical

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

Inferences
1. 25 percentile value of sat and sun is more than that of thursday and friday
2. same goes for 75 percentile
3. iqr for weekends are higher than that of weekdays
4. if iqr is high we have high paying capacities on weekends

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex',split=True)
plt.show()
```

# Regplot
